```python
#Import dependencies
import pandas as pd
```

```python
# Assign a variable to each 2017 Q1 csv file
jan_2017_csv = "Data/Full csv files/201701-citibike-tripdata.csv"
feb_2017_csv = "Data/Full csv files/201702-citibike-tripdata.csv"
mar_2017_csv = "Data/Full csv files/201703-citibike-tripdata.csv"

# Assign a variable to each 2018 Q1 csv file
jan_2018_csv = "Data/Full csv files/201801-citibike-tripdata.csv"
feb_2018_csv = "Data/Full csv files/201802-citibike-tripdata.csv"
mar_2018_csv = "Data/Full csv files/201803-citibike-tripdata.csv"

# Read 2017 csv data files with pandas and store in a dataframe
jan_2017_df = pd.read_csv(
    jan_2017_csv, encoding="utf-8", low_memory=False)

feb_2017_df = pd.read_csv(
    feb_2017_csv, encoding="utf-8", low_memory=False)

mar_2017_df = pd.read_csv(
    mar_2017_csv, encoding="utf-8", low_memory=False)

# Read 2018 csv data files with pandas and store in a dataframe
jan_2018_df = pd.read_csv(
    jan_2018_csv, encoding="utf-8", low_memory=False)

feb_2018_df = pd.read_csv(
    feb_2018_csv, encoding="utf-8", low_memory=False)

mar_2018_df = pd.read_csv(
    mar_2018_csv, encoding="utf-8", low_memory=False)


```

```python
#Combine all of the dataframes into one for q1 2017

jan_feb = jan_2017_df.append(feb_2017_df)

q1_2017 = jan_feb.append(mar_2017_df)

#Combine all of the dataframes into one for q1 2018

jan_feb = jan_2018_df.append(feb_2018_df)

q1_2018 = jan_feb.append(mar_2018_df)
```

```python
#Drop all of the NA values
q1_2017 = q1_2017.dropna()

#Drop all of the NA values
q1_2018 = q1_2018.dropna()
```

```python
#Add a new column that just contains the year since we don't care about exact dates/times

q1_2017['Year'] = 2017

q1_2018['Year'] = 2018
q1_2018.head()
```

```python
#Drop columns that are not needed for analysis

q1_2017 = q1_2017.drop(['Start Time', 'Stop Time'], axis=1)

q1_2018 = q1_2018.drop(['starttime', 'stoptime'], axis=1)
```

```python
#Rename the columns in the 2018 file so that they match 2017

q1_2018 = q1_2018.rename({'tripduration': 'Trip Duration', 'start station id': 'Start Station ID', 
                'start station name': 'Start Station Name', 'start station latitude' : 'Start Station Latitude', 
                 'start station longitude': 'Start Station Longitude', 'end station id': 'End Station ID', 
                'end station name': 'End Station Name', 'end station latitude': 'End Station Latitude', 
                'end station longitude': 'End Station Longitude', 'bikeid': 'Bike ID', 'usertype': 'User Type',
               'birth year': 'Birth Year', 'gender': 'Gender'}, axis='columns')
```

```python
counts = q1_2018.count()
counts
```

```python
#Combine the 2017 and 2018 data into one dataframe
q1_2017_2018 = q1_2017.append(q1_2018, sort=False)

```

```python
total_count = q1_2017_2018.count()
total_count
```

```python
#Change the birth year, gender, station id and trip duration columns to integers

q1_2017_2018["Birth Year"] = q1_2017_2018["Birth Year"].astype(int)

q1_2017_2018["Gender"] = q1_2017_2018["Gender"].astype(int)

q1_2017_2018["Start Station ID"] = q1_2017_2018["Start Station ID"].astype(int)

q1_2017_2018["End Station ID"] = q1_2017_2018["End Station ID"].astype(int)

q1_2017_2018["Trip Duration"] = q1_2017_2018["Trip Duration"].astype(int)

#Now change the station ids to strings so Tableau does not consider them measures

q1_2017_2018["Start Station ID"] = q1_2017_2018["Start Station ID"].astype(str)
q1_2017_2018["End Station ID"] = q1_2017_2018["End Station ID"].astype(str)
```

```python
#Replace the 0, 1, 2 values in the Gender column with unknown, male, female

q1_2017_2018['Gender'].replace({0:'Unknown', 1:'Male', 2:'Female'}, inplace=True)
```
